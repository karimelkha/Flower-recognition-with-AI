```
model = tf.keras.Sequential()
inputs = Input ( shape = (  150 , 150 , 3 ))
# R√©utilisation de VGG16
vgg16_features = keras.applications.vgg16.VGG16( include_top=False , weights= "imagenet",input_tensor=inputs)
vgg16_features.trainable= False


model.add(vgg16_features)
model.add(Flatten())
model.add(tf.keras.layers.Dense(512,activation='relu'))
# Le nombre de labels doit correspondre au nombre pass√© en param√®tre de la couche d'activation de sortie
model.add(tf.keras.layers.Dense(5,activation='softmax'))

dataset = tf.keras.preprocessing.image_dataset_from_directory("flowers/",image_size=(150,150))

model.compile(loss="categorical_crossentropy", optimizer=keras.optimizers.SGD(learning_rate=1e-4, momentum = 0.9))



train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        rescale=1./150,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./150)

train_generator = train_datagen.flow_from_directory(
        'flowers',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
validation_generator = test_datagen.flow_from_directory(
        'flowers_test',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')

model.fit(
        train_generator,
        steps_per_epoch=20,
        epochs=10,
        validation_data=validation_generator,
        validation_steps=10)
```
